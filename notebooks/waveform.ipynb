```python
from pycbc.waveform import td_approximants, fd_approximants, get_td_waveform
import numpy as np
import matplotlib.pyplot as plt

# List of td approximants that are available
print(td_approximants())
```

```python
hp, hc = get_td_waveform(approximant='IMRPhenomPv3',
                                 mass1=100,
                                 mass2=80,
                                 spin1z=0.6,
                                 delta_t=1.0/10000,
                                 f_lower=25)
```

```python
fig,ax = plt.subplots(figsize=(16,5))
ax.plot(hp.sample_times, hp,color="#08519c",lw=2)
ax.set_xlim(-0.5,.1)
ax.axis('off')
plt.savefig('waveform.png',dpi=200)
```
